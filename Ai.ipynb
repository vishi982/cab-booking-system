```python
import yfinance as yf
import pandas as pd

# Simulate user-inputted stock tickers
tickers = ['AAPL', 'MSFT']

# Define the period for historical data
period = '5y'

# Dictionary to store historical price data
historical_price_data = {}

# Dictionary to store financial metrics
financial_metrics = {}

for ticker in tickers:
    # Fetch historical price data
    stock = yf.Ticker(ticker)
    hist = stock.history(period=period)
    historical_price_data[ticker] = hist

    # Fetch financial metrics
    # yfinance provides some fundamental data in the 'info' attribute
    info = stock.info
    metrics = {
        'Beta': info.get('beta'),
        'PE Ratio': info.get('trailingPE'),
        'Debt to Equity': info.get('debtToEquity'),
        'Industry': info.get('industry'),
        'Sector': info.get('sector')
    }
    financial_metrics[ticker] = metrics

# Convert dictionaries to DataFrames for structured storage
historical_price_df = pd.concat(historical_price_data, axis=1)
financial_metrics_df = pd.DataFrame.from_dict(financial_metrics, orient='index')

# Display the first few rows of the historical price data
print("Historical Price Data:")
display(historical_price_df.head())

# Display the financial metrics
print("\nFinancial Metrics:")
display(financial_metrics_df)
```

## Portfolio scoring logic

### Subtask:
Develop the code to calculate the portfolio score based on the defined criteria and acquired data. This will likely involve analyzing the individual stocks and their combined effect on the overall portfolio.


**Reasoning**:
Develop the code to calculate the portfolio score based on the defined criteria and acquired data, including annualized return, Sharpe Ratio, weighted averages for Beta, P/E, and Debt-to-Equity, a diversification score, normalization of metrics, and calculation of the composite score.



```python
import numpy as np
from scipy.stats import gmean

def calculate_portfolio_score(historical_price_df, financial_metrics_df, weights):
    """
    Calculates a composite portfolio score based on various financial metrics.

    Args:
        historical_price_df (pd.DataFrame): DataFrame with historical price data
                                             for each stock, indexed by Date.
        financial_metrics_df (pd.DataFrame): DataFrame with financial metrics
                                            for each stock, indexed by ticker.
        weights (dict): A dictionary where keys are stock tickers and values
                        are their respective portfolio weights (summing to 1).

    Returns:
        tuple: A tuple containing:
            - dict: Individual normalized metric scores.
            - float: Composite portfolio score.
    """
    # Ensure weights sum to 1
    total_weight = sum(weights.values())
    if not np.isclose(total_weight, 1.0):
        print(f"Warning: Portfolio weights do not sum to 1. They sum to {total_weight}. Normalizing weights.")
        weights = {ticker: weight / total_weight for ticker, weight in weights.items()}


    # 2. Calculate annualized return for each stock
    annualized_returns = {}
    for ticker in historical_price_df.columns.get_level_values(0).unique():
        if ticker in weights and weights[ticker] > 0: # Only calculate for stocks in the portfolio
            close_prices = historical_price_df[ticker]['Close'].dropna()
            if len(close_prices) > 1:
                # Calculate daily returns
                daily_returns = close_prices.pct_change().dropna()
                # Calculate annualized return (assuming ~252 trading days in a year)
                annualized_return = (1 + daily_returns.mean())**252 - 1
                annualized_returns[ticker] = annualized_return
            else:
                annualized_returns[ticker] = 0 # Or handle as needed

    # 3. Calculate Sharpe Ratio for each stock
    risk_free_rate = 0.02  # Placeholder risk-free rate (2%)
    sharpe_ratios = {}
    for ticker in historical_price_df.columns.get_level_values(0).unique():
         if ticker in weights and weights[ticker] > 0: # Only calculate for stocks in the portfolio
            close_prices = historical_price_df[ticker]['Close'].dropna()
            if len(close_prices) > 1:
                daily_returns = close_prices.pct_change().dropna()
                if daily_returns.std() > 0:
                     # Assuming daily risk-free rate is risk_free_rate / 252
                    sharpe_ratio = (daily_returns.mean() - (risk_free_rate / 252)) / daily_returns.std() * np.sqrt(252)
                    sharpe_ratios[ticker] = sharpe_ratio
                else:
                    sharpe_ratios[ticker] = 0 # Handle cases with no volatility
            else:
                 sharpe_ratios[ticker] = 0 # Handle cases with insufficient data


    # 4. Calculate weighted average Beta, P/E Ratio, and Debt-to-Equity
    weighted_beta = 0
    weighted_pe_ratio = 0
    weighted_debt_to_equity = 0
    valid_weights_sum = 0 # To handle missing financial data

    for ticker, weight in weights.items():
        if ticker in financial_metrics_df.index:
            metrics = financial_metrics_df.loc[ticker]
            if pd.notna(metrics['Beta']):
                weighted_beta += metrics['Beta'] * weight
                valid_weights_sum += weight # Only sum weights for stocks with Beta

            if pd.notna(metrics['PE Ratio']):
                weighted_pe_ratio += metrics['PE Ratio'] * weight

            if pd.notna(metrics['Debt to Equity']):
                weighted_debt_to_equity += metrics['Debt to Equity'] * weight

    # Adjust weighted beta if some stocks were missing Beta
    if valid_weights_sum > 0 and not np.isclose(valid_weights_sum, 1.0):
         weighted_beta /= valid_weights_sum
         weighted_pe_ratio /= sum(weights.values()) # Assuming PE and D/E are available for all or handled by fillna
         weighted_debt_to_equity /= sum(weights.values()) # Assuming PE and D/E are available for all or handled by fillna


    # 5. Calculate diversification score (using unique industries/sectors count)
    industries = set()
    sectors = set()
    for ticker, weight in weights.items():
         if ticker in financial_metrics_df.index and weight > 0:
            industry = financial_metrics_df.loc[ticker, 'Industry']
            sector = financial_metrics_df.loc[ticker, 'Sector']
            if pd.notna(industry):
                industries.add(industry)
            if pd.notna(sector):
                sectors.add(sector)

    diversification_score_raw = len(industries) + len(sectors) # Simple count


    # Calculate weighted average annualized return and Sharpe Ratio for the portfolio
    weighted_annualized_return = sum(annualized_returns.get(ticker, 0) * weights[ticker] for ticker in weights)
    weighted_sharpe_ratio = sum(sharpe_ratios.get(ticker, 0) * weights[ticker] for ticker in weights)


    # 6. Normalize or scale the calculated metrics
    # Define scaling ranges or methods. Simple min-max scaling for now.
    # Need to handle cases where min == max for a metric.
    # Also need to consider if a higher or lower value is better.

    raw_scores = {
        'Annualized Return': weighted_annualized_return,
        'Sharpe Ratio': weighted_sharpe_ratio,
        'Beta': weighted_beta,
        'PE Ratio': weighted_pe_ratio,
        'Debt to Equity': weighted_debt_to_equity,
        'Diversification': diversification_score_raw
    }

    normalized_scores = {}
    # Define how to normalize each metric (higher is better vs lower is better)
    # For simplicity, we'll define hypothetical min/max ranges or use observed min/max.
    # In a real application, these ranges might be based on benchmarks or industry standards.

    # Observed min/max for normalization (handle potential division by zero)
    observed_min_max = {}
    for metric, value in raw_scores.items():
         # Collect all individual stock values for return and sharpe for normalization
         if metric == 'Annualized Return':
              values = list(annualized_returns.values())
         elif metric == 'Sharpe Ratio':
             values = list(sharpe_ratios.values())
         elif metric in ['Beta', 'PE Ratio', 'Debt to Equity']:
             values = financial_metrics_df[metric].dropna().tolist()
         elif metric == 'Diversification':
              # Diversification score is already a portfolio-level metric
              values = [value]
         else:
              values = [value] # Should not happen with current metrics

         if values:
            min_val = min(values)
            max_val = max(values)
            # Add portfolio weighted average to the min/max for consistency if needed,
            # or just use the portfolio value for min/max if it's a portfolio metric
            if metric not in ['Annualized Return', 'Sharpe Ratio']:
                 min_val = min(min_val, value)
                 max_val = max(max_val, value)

            observed_min_max[metric] = {'min': min_val, 'max': max_val}
         else:
             observed_min_max[metric] = {'min': 0, 'max': 1} # Default range if no data


    # Normalization (0-100 scale)
    for metric, value in raw_scores.items():
        min_val = observed_min_max[metric]['min']
        max_val = observed_min_max[metric]['max']

        if max_val == min_val:
            normalized_scores[metric] = 50 # Assign a neutral score if no variation
        else:
            # Metrics where higher is better
            if metric in ['Annualized Return', 'Sharpe Ratio', 'Diversification']:
                normalized_scores[metric] = ((value - min_val) / (max_val - min_val)) * 100
            # Metrics where lower is better (invert the scale)
            elif metric in ['Beta', 'PE Ratio', 'Debt to Equity']:
                 normalized_scores[metric] = 100 - (((value - min_val) / (max_val - min_val)) * 100)
            else:
                 normalized_scores[metric] = 0 # Should not happen


    # 7. Apply predefined weights to the normalized scores
    # Example weighting scheme (summing to 1) - these can be adjusted
    criterion_weights = {
        'Annualized Return': 0.30,
        'Sharpe Ratio': 0.20,
        'Beta': 0.25,
        'PE Ratio': 0.05,
        'Debt to Equity': 0.05,
        'Diversification': 0.15
    }

    composite_score = 0
    for metric, weight in criterion_weights.items():
        if metric in normalized_scores:
            composite_score += normalized_scores[metric] * weight
        else:
            print(f"Warning: Metric '{metric}' not found in normalized scores.")


    # Ensure the composite score is within a reasonable range (0-100)
    composite_score = max(0, min(100, composite_score))


    # 8. Return individual metric scores and the composite portfolio score
    return normalized_scores, composite_score

# Example Usage:
# Assume an equal-weighted portfolio for the stocks in financial_metrics_df
portfolio_tickers = financial_metrics_df.index.tolist()
portfolio_weights = {ticker: 1/len(portfolio_tickers) for ticker in portfolio_tickers}

individual_scores, composite_portfolio_score = calculate_portfolio_score(
    historical_price_df, financial_metrics_df, portfolio_weights
)

print("\nIndividual Normalized Metric Scores:")
for metric, score in individual_scores.items():
    print(f"- {metric}: {score:.2f}")

print(f"\nComposite Portfolio Score: {composite_portfolio_score:.2f}")
```

## Stock replacement recommendations

### Subtask:
Based on the scoring, develop logic to suggest which stocks might be replaced to improve the portfolio score. This could involve identifying underperforming stocks or suggesting alternatives that better fit the scoring criteria.


**Reasoning**:
Define a function to generate stock replacement recommendations based on low-scoring metrics and then call this function and print the results.



```python
def suggest_replacements(individual_scores, financial_metrics_df, historical_price_df):
    """
    Suggests stock replacements to improve the portfolio score based on low-scoring metrics.

    Args:
        individual_scores (dict): Dictionary of individual normalized metric scores (0-100).
        financial_metrics_df (pd.DataFrame): DataFrame with financial metrics
                                            for each stock, indexed by ticker.
        historical_price_df (pd.DataFrame): DataFrame with historical price data
                                             for each stock, indexed by Date.

    Returns:
        list: A list of strings, each suggesting a potential recommendation.
    """
    recommendations = []
    low_score_threshold = 60 # Define a threshold for considering a score 'low'

    # Identify low-scoring metrics
    low_scoring_metrics = {
        metric: score for metric, score in individual_scores.items()
        if score < low_score_threshold
    }

    # Determine characteristics of potentially better-performing stocks
    # and simulate suggestions based on low-scoring metrics

    # Example: Low Annualized Return or Sharpe Ratio
    if 'Annualized Return' in low_scoring_metrics or 'Sharpe Ratio' in low_scoring_metrics:
        recommendations.append("Consider replacing stocks with low historical returns or Sharpe Ratios with those exhibiting stronger historical performance.")

    # Example: High Beta (low Beta score)
    if 'Beta' in low_scoring_metrics:
        recommendations.append("If your Beta score is low (indicating high portfolio risk), consider replacing high-Beta stocks with those that have lower Beta.")

    # Example: High P/E Ratio or Debt-to-Equity (low scores)
    if 'PE Ratio' in low_scoring_metrics or 'Debt to Equity' in low_scoring_metrics:
        recommendations.append("If your Fundamentals scores are low (indicating high PE or Debt), explore companies with more favorable valuation (lower PE) or stronger balance sheets (lower Debt-to-Equity).")

    # Example: Low Diversification score
    if 'Diversification' in low_scoring_metrics:
        # Identify current industries/sectors to suggest alternatives
        current_industries = financial_metrics_df['Industry'].dropna().unique().tolist()
        current_sectors = financial_metrics_df['Sector'].dropna().unique().tolist()
        recommendations.append(f"Your diversification score is low. Consider adding stocks from different industries (current: {', '.join(current_industries)}) or sectors (current: {', '.join(current_sectors)}) to improve diversification.")

    # Add a general recommendation if no specific low scores were found but score isn't perfect
    if not recommendations and sum(individual_scores.values())/len(individual_scores) < 100:
         recommendations.append("Your portfolio metrics are relatively balanced, but there's always room for improvement. Continuously review your holdings against your investment goals.")
    elif not recommendations:
         recommendations.append("Your portfolio scores highly across all metrics! Keep monitoring your investments.")


    # In a real application, this is where you would query a database of stocks
    # based on the desired characteristics (e.g., search for stocks with Beta < 1,
    # or in a specific sector not currently in the portfolio).
    # For this simulation, we just provide generic suggestions.

    return recommendations

# Call the function with the calculated scores and dataframes
replacement_recommendations = suggest_replacements(individual_scores, financial_metrics_df, historical_price_df)

# Print the recommendations
print("\nStock Replacement Recommendations:")
if replacement_recommendations:
    for rec in replacement_recommendations:
        print(f"- {rec}")
else:
    print("No specific recommendations generated based on current criteria.")
```

```python
from flask import Flask, render_template, request, jsonify
import pandas as pd
import yfinance as yf
import numpy as np
from scipy.stats import gmean

# Initialize Flask app
app = Flask(__name__)

# --- Placeholder for previously defined functions ---
# In a real application, these functions would be imported or defined here
# For this example, we will include their definitions.

def calculate_portfolio_score(historical_price_df, financial_metrics_df, weights):
    """
    Calculates a composite portfolio score based on various financial metrics.
    (Includes code from the previous subtask)
    """
    # Ensure weights sum to 1
    total_weight = sum(weights.values())
    if not np.isclose(total_weight, 1.0):
        weights = {ticker: weight / total_weight for ticker, weight in weights.items()}

    # Calculate annualized return for each stock
    annualized_returns = {}
    for ticker in historical_price_df.columns.get_level_values(0).unique():
        if ticker in weights and weights[ticker] > 0:
            close_prices = historical_price_df[ticker]['Close'].dropna()
            if len(close_prices) > 1:
                daily_returns = close_prices.pct_change().dropna()
                annualized_return = (1 + daily_returns.mean())**252 - 1
                annualized_returns[ticker] = annualized_return
            else:
                annualized_returns[ticker] = 0

    # Calculate Sharpe Ratio for each stock
    risk_free_rate = 0.02
    sharpe_ratios = {}
    for ticker in historical_price_df.columns.get_level_values(0).unique():
         if ticker in weights and weights[ticker] > 0:
            close_prices = historical_price_df[ticker]['Close'].dropna()
            if len(close_prices) > 1:
                daily_returns = close_prices.pct_change().dropna()
                if daily_returns.std() > 0:
                    sharpe_ratio = (daily_returns.mean() - (risk_free_rate / 252)) / daily_returns.std() * np.sqrt(252)
                    sharpe_ratios[ticker] = sharpe_ratio
                else:
                    sharpe_ratios[ticker] = 0
            else:
                 sharpe_ratios[ticker] = 0

    # Calculate weighted average Beta, P/E Ratio, and Debt-to-Equity
    weighted_beta = 0
    weighted_pe_ratio = 0
    weighted_debt_to_equity = 0
    valid_weights_sum = 0

    for ticker, weight in weights.items():
        if ticker in financial_metrics_df.index:
            metrics = financial_metrics_df.loc[ticker]
            if pd.notna(metrics['Beta']):
                weighted_beta += metrics['Beta'] * weight
                valid_weights_sum += weight

            if pd.notna(metrics['PE Ratio']):
                weighted_pe_ratio += metrics['PE Ratio'] * weight

            if pd.notna(metrics['Debt to Equity']):
                weighted_debt_to_equity += metrics['Debt to Equity'] * weight

    if valid_weights_sum > 0 and not np.isclose(valid_weights_sum, 1.0):
         weighted_beta /= valid_weights_sum
         # Assuming PE and D/E are available for all or handled by fillna
         weighted_pe_ratio /= sum(weights.values())
         weighted_debt_to_equity /= sum(weights.values())

    # Calculate diversification score
    industries = set()
    sectors = set()
    for ticker, weight in weights.items():
         if ticker in financial_metrics_df.index and weight > 0:
            industry = financial_metrics_df.loc[ticker, 'Industry']
            sector = financial_metrics_df.loc[ticker, 'Sector']
            if pd.notna(industry):
                industries.add(industry)
            if pd.notna(sector):
                sectors.add(sector)

    diversification_score_raw = len(industries) + len(sectors)

    # Calculate weighted average annualized return and Sharpe Ratio for the portfolio
    weighted_annualized_return = sum(annualized_returns.get(ticker, 0) * weights[ticker] for ticker in weights)
    weighted_sharpe_ratio = sum(sharpe_ratios.get(ticker, 0) * weights[ticker] for ticker in weights)

    # Normalize or scale the calculated metrics
    raw_scores = {
        'Annualized Return': weighted_annualized_return,
        'Sharpe Ratio': weighted_sharpe_ratio,
        'Beta': weighted_beta,
        'PE Ratio': weighted_pe_ratio,
        'Debt to Equity': weighted_debt_to_equity,
        'Diversification': diversification_score_raw
    }

    normalized_scores = {}
    observed_min_max = {}
    for metric, value in raw_scores.items():
         if metric == 'Annualized Return':
              values = list(annualized_returns.values())
         elif metric == 'Sharpe Ratio':
             values = list(sharpe_ratios.values())
         elif metric in ['Beta', 'PE Ratio', 'Debt to Equity']:
             values = financial_metrics_df[metric].dropna().tolist()
         elif metric == 'Diversification':
              values = [value]
         else:
              values = [value]

         if values:
            min_val = min(values)
            max_val = max(values)
            if metric not in ['Annualized Return', 'Sharpe Ratio']:
                 min_val = min(min_val, value)
                 max_val = max(max_val, value)

            observed_min_max[metric] = {'min': min_val, 'max': max_val}
         else:
             observed_min_max[metric] = {'min': 0, 'max': 1}

    for metric, value in raw_scores.items():
        min_val = observed_min_max[metric]['min']
        max_val = observed_min_max[metric]['max']

        if max_val == min_val:
            normalized_scores[metric] = 50
        else:
            if metric in ['Annualized Return', 'Sharpe Ratio', 'Diversification']:
                normalized_scores[metric] = ((value - min_val) / (max_val - min_val)) * 100
            elif metric in ['Beta', 'PE Ratio', 'Debt to Equity']:
                 normalized_scores[metric] = 100 - (((value - min_val) / (max_val - min_val)) * 100)
            else:
                 normalized_scores[metric] = 0

    criterion_weights = {
        'Annualized Return': 0.30,
        'Sharpe Ratio': 0.20,
        'Beta': 0.25,
        'PE Ratio': 0.05,
        'Debt to Equity': 0.05,
        'Diversification': 0.15
    }

    composite_score = 0
    for metric, weight in criterion_weights.items():
        if metric in normalized_scores:
            composite_score += normalized_scores[metric] * weight

    composite_score = max(0, min(100, composite_score))

    return normalized_scores, composite_score

def suggest_replacements(individual_scores, financial_metrics_df, historical_price_df):
    """
    Suggests stock replacements to improve the portfolio score based on low-scoring metrics.
    (Includes code from the previous subtask)
    """
    recommendations = []
    low_score_threshold = 60

    low_scoring_metrics = {
        metric: score for metric, score in individual_scores.items()
        if score < low_score_threshold
    }

    if 'Annualized Return' in low_scoring_metrics or 'Sharpe Ratio' in low_scoring_metrics:
        recommendations.append("Consider replacing stocks with low historical returns or Sharpe Ratios with those exhibiting stronger historical performance.")

    if 'Beta' in low_scoring_metrics:
        recommendations.append("If your Beta score is low (indicating high portfolio risk), consider replacing high-Beta stocks with those that have lower Beta.")

    if 'PE Ratio' in low_scoring_metrics or 'Debt to Equity' in low_scoring_metrics:
        recommendations.append("If your Fundamentals scores are low (indicating high PE or Debt), explore companies with more favorable valuation (lower PE) or stronger balance sheets (lower Debt-to-Equity).")

    if 'Diversification' in low_scoring_metrics:
        current_industries = financial_metrics_df['Industry'].dropna().unique().tolist()
        current_sectors = financial_metrics_df['Sector'].dropna().unique().tolist()
        recommendations.append(f"Your diversification score is low. Consider adding stocks from different industries (current: {', '.join(current_industries)}) or sectors (current: {', '.join(current_sectors)}) to improve diversification.")

    if not recommendations and sum(individual_scores.values())/len(individual_scores) < 100:
         recommendations.append("Your portfolio metrics are relatively balanced, but there's always room for improvement. Continuously review your holdings against your investment goals.")
    elif not recommendations:
         recommendations.append("Your portfolio scores highly across all metrics! Keep monitoring your investments.")

    return recommendations


# --- Flask Routes ---

@app.route('/')
def index():
    """
    Renders the main page with the portfolio input form.
    """
    return render_template('index.html')

@app.route('/calculate_score', methods=['POST'])
def calculate_score():
    """
    Receives portfolio data, calculates score and recommendations, and returns results.
    """
    data = request.json # Assuming data is sent as JSON
    portfolio_data = data.get('portfolio', [])

    if not portfolio_data:
        return jsonify({"error": "No portfolio data provided"}), 400

    tickers = [item['ticker'].upper() for item in portfolio_data]
    weights = {item['ticker'].upper(): float(item['weight']) for item in portfolio_data}

    # --- Data Acquisition ---
    # Fetch historical price data and financial metrics for the tickers
    # This part is simplified for the example. In a real app, handle errors and caching.
    historical_price_data = {}
    financial_metrics = {}
    period = '5y' # Define the period for historical data

    for ticker in tickers:
        try:
            stock = yf.Ticker(ticker)
            hist = stock.history(period=period)
            historical_price_data[ticker] = hist

            info = stock.info
            metrics = {
                'Beta': info.get('beta'),
                'PE Ratio': info.get('trailingPE'),
                'Debt to Equity': info.get('debtToEquity'),
                'Industry': info.get('industry'),
                'Sector': info.get('sector')
            }
            financial_metrics[ticker] = metrics
        except Exception as e:
            print(f"Error fetching data for {ticker}: {e}")
            # Handle missing data for a ticker, e.g., skip or assign default values
            financial_metrics[ticker] = {} # Assign empty dict if data fetching fails
            historical_price_data[ticker] = pd.DataFrame() # Assign empty DataFrame


    # Filter out tickers where data fetching failed before creating DataFrames
    valid_tickers = [t for t in tickers if not historical_price_data[t].empty and financial_metrics[t]]
    if not valid_tickers:
        return jsonify({"error": "Could not fetch data for any of the provided tickers."}), 400

    # Recreate weights dictionary with only valid tickers and re-normalize
    valid_weights = {ticker: weights[ticker] for ticker in valid_tickers}
    total_valid_weight = sum(valid_weights.values())
    if not np.isclose(total_valid_weight, 1.0):
        valid_weights = {ticker: weight / total_valid_weight for ticker, weight in valid_weights.items()}


    historical_price_df = pd.concat({t: historical_price_data[t] for t in valid_tickers}, axis=1)
    financial_metrics_df = pd.DataFrame.from_dict({t: financial_metrics[t] for t in valid_tickers}, orient='index')


    # --- Calculate Score and Recommendations ---
    individual_scores, composite_portfolio_score = calculate_portfolio_score(
        historical_price_df, financial_metrics_df, valid_weights
    )

    replacement_recommendations = suggest_replacements(
        individual_scores, financial_metrics_df, historical_price_df
    )

    # --- Return Results ---
    results = {
        "composite_score": round(composite_portfolio_score, 2),
        "individual_scores": {metric: round(score, 2) for metric, score in individual_scores.items()},
        "recommendations": replacement_recommendations
    }

    return jsonify(results)


# --- HTML Template (index.html) ---
# Create a string that represents the HTML content.
# In a real Flask app, this would be in a separate file in a 'templates' folder.
html_template = """
<!doctype html>
<html lang="en">
<head>
    <meta charset="utf-8">
    <meta name="viewport" content="width=device-width, initial-scale=1, shrink-to-fit=no">
    <title>Portfolio Scorer</title>
    <style>
        body { font-family: Arial, sans-serif; line-height: 1.6; margin: 20px; }
        .container { max-width: 800px; margin: auto; background: #f9f9f9; padding: 20px; border-radius: 8px; }
        h1, h2 { text-align: center; color: #333; }
        .form-group { margin-bottom: 15px; }
        label { display: block; margin-bottom: 5px; font-weight: bold; }
        input[type="text"], input[type="number"] { width: calc(100% - 22px); padding: 10px; border: 1px solid #ddd; border-radius: 4px; }
        #portfolio-items { border: 1px solid #eee; padding: 10px; border-radius: 4px; }
        .portfolio-item { display: flex; justify-content: space-between; align-items: center; margin-bottom: 10px; padding: 10px; background: #fff; border: 1px solid #ddd; border-radius: 4px; }
        .portfolio-item input { margin-right: 10px; flex-grow: 1; }
        .portfolio-item button { background-color: #dc3545; color: white; border: none; padding: 5px 10px; border-radius: 4px; cursor: pointer; }
        .portfolio-item button:hover { background-color: #c82333; }
        #add-stock-btn { display: block; width: 100%; padding: 10px; background-color: #28a745; color: white; border: none; border-radius: 4px; cursor: pointer; font-size: 16px; }
        #add-stock-btn:hover { background-color: #218838; }
        button[type="submit"] { display: block; width: 100%; padding: 10px; background-color: #007bff; color: white; border: none; border-radius: 4px; cursor: pointer; font-size: 18px; margin-top: 20px; }
        button[type="submit"]:hover { background-color: #0056b3; }
        #results { margin-top: 30px; padding: 20px; background: #e9e9e9; border-radius: 8px; display: none; }
        #results h2 { color: #007bff; }
        #scores-list { list-style: none; padding: 0; }
        #scores-list li { background: #fff; padding: 10px; margin-bottom: 8px; border-radius: 4px; border: 1px solid #ddd; }
        #recommendations-list { list-style: disc; margin-left: 20px; }
        #recommendations-list li { margin-bottom: 8px; }
        .error { color: red; text-align: center; margin-top: 15px; }
    </style>
</head>
<body>
    <div class="container">
        <h1>Portfolio Scorer</h1>
        <form id="portfolio-form">
            <div id="portfolio-items">
                <div class="portfolio-item">
                    <input type="text" class="ticker" placeholder="Stock Ticker (e.g., AAPL)" required>
                    <input type="number" class="weight" placeholder="Weight % (e.g., 50)" min="1" max="100" required>
                    <button type="button" class="remove-stock-btn">Remove</button>
                </div>
            </div>
            <button type="button" id="add-stock-btn">Add Another Stock</button>
            <button type="submit">Calculate Score</button>
            <div id="form-error" class="error"></div>
        </form>

        <div id="results">
            <h2>Portfolio Score: <span id="composite-score">-</span></h2>
            <h3>Individual Metric Scores:</h3>
            <ul id="scores-list">
                <!-- Scores will be populated here by JavaScript -->
            </ul>
            <h3>Recommendations:</h3>
            <ul id="recommendations-list">
                 <!-- Recommendations will be populated here by JavaScript -->
            </ul>
             <div id="results-error" class="error"></div>
        </div>
    </div>

    <script>
        document.addEventListener('DOMContentLoaded', function() {
            const portfolioItemsDiv = document.getElementById('portfolio-items');
            const addStockBtn = document.getElementById('add-stock-btn');
            const portfolioForm = document.getElementById('portfolio-form');
            const resultsDiv = document.getElementById('results');
            const compositeScoreSpan = document.getElementById('composite-score');
            const scoresList = document.getElementById('scores-list');
            const recommendationsList = document.getElementById('recommendations-list');
            const formErrorDiv = document.getElementById('form-error');
            const resultsErrorDiv = document.getElementById('results-error');


            // Add stock item functionality
            addStockBtn.addEventListener('click', function() {
                const newItem = document.createElement('div');
                newItem.classList.add('portfolio-item');
                newItem.innerHTML = `
                    <input type="text" class="ticker" placeholder="Stock Ticker (e.g., MSFT)" required>
                    <input type="number" class="weight" placeholder="Weight % (e.g., 50)" min="1" max="100" required>
                    <button type="button" class="remove-stock-btn">Remove</button>
                `;
                portfolioItemsDiv.appendChild(newItem);
            });

            // Remove stock item functionality using event delegation
            portfolioItemsDiv.addEventListener('click', function(event) {
                if (event.target.classList.contains('remove-stock-btn')) {
                    // Ensure there's at least one item left
                    if (portfolioItemsDiv.children.length > 1) {
                        event.target.parentElement.remove();
                    } else {
                        alert("You must have at least one stock in your portfolio.");
                    }
                }
            });

            // Handle form submission
            portfolioForm.addEventListener('submit', function(event) {
                event.preventDefault(); // Prevent default form submission

                formErrorDiv.textContent = ''; // Clear previous form errors
                resultsErrorDiv.textContent = ''; // Clear previous results errors
                resultsDiv.style.display = 'none'; // Hide previous results

                const portfolioItems = portfolioItemsDiv.querySelectorAll('.portfolio-item');
                const portfolioData = [];
                let totalWeight = 0;

                portfolioItems.forEach(item => {
                    const tickerInput = item.querySelector('.ticker');
                    const weightInput = item.querySelector('.weight');
                    const ticker = tickerInput.value.trim();
                    const weight = parseFloat(weightInput.value);

                    if (ticker && !isNaN(weight) && weight > 0) {
                        portfolioData.push({ ticker: ticker, weight: weight });
                        totalWeight += weight;
                    } else {
                        // Basic validation
                        formErrorDiv.textContent = "Please enter valid ticker and weight for all stocks.";
                        return; // Stop processing if validation fails
                    }
                });

                 // Check if total weight is close to 100
                if (Math.abs(totalWeight - 100) > 0.1) { // Allow for slight floating point inaccuracies
                     formErrorDiv.textContent = `Total weight must be 100%. Current total: ${totalWeight.toFixed(2)}%.`;
                     return; // Stop if weights don't sum to 100
                 }


                if (portfolioData.length === 0) {
                    formErrorDiv.textContent = "Please add at least one stock to your portfolio.";
                    return;
                }

                // Send data to the backend
                fetch('/calculate_score', {
                    method: 'POST',
                    headers: {
                        'Content-Type': 'application/json'
                    },
                    body: JSON.stringify({ portfolio: portfolioData })
                })
                .then(response => response.json().then(data => ({ status: response.status, body: data })))
                .then(result => {
                    if (result.status === 200) {
                        const data = result.body;
                        // Display results
                        compositeScoreSpan.textContent = data.composite_score;

                        // Display individual scores
                        scoresList.innerHTML = ''; // Clear previous scores
                        for (const [metric, score] of Object.entries(data.individual_scores)) {
                            const li = document.createElement('li');
                            li.textContent = `${metric}: ${score}`;
                            scoresList.appendChild(li);
                        }

                        // Display recommendations
                        recommendationsList.innerHTML = ''; // Clear previous recommendations
                         if (data.recommendations && data.recommendations.length > 0) {
                            data.recommendations.forEach(rec => {
                                const li = document.createElement('li');
                                li.textContent = rec;
                                recommendationsList.appendChild(li);
                            });
                        } else {
                             const li = document.createElement('li');
                             li.textContent = "No specific recommendations at this time.";
                             recommendationsList.appendChild(li);
                        }


                        resultsDiv.style.display = 'block'; // Show results section
                    } else {
                        // Handle backend errors
                        resultsErrorDiv.textContent = result.body.error || 'An error occurred while calculating the score.';
                         resultsDiv.style.display = 'block'; // Show results section even with error
                    }
                })
                .catch(error => {
                    console.error('Error:', error);
                    resultsErrorDiv.textContent = 'An unexpected error occurred.';
                     resultsDiv.style.display = 'block'; // Show results section even with error
                });
            });
        });
    </script>
</body>
</html>
"""

# Save the HTML template to a file in a 'templates' folder
# In a real Flask app, you would manage this file separately.
# For this example within a notebook, we simulate saving it.
import os

# Create a 'templates' directory if it doesn't exist
if not os.path.exists('templates'):
    os.makedirs('templates')

# Write the HTML content to index.html inside the 'templates' directory
with open('templates/index.html', 'w') as f:
    f.write(html_template)

print("Flask app and HTML template created.")
print("To run the Flask app, you would typically use:")
print("from flask import Flask")
print("...")
print("if __name__ == '__main__':")
print("    app.run(debug=True)")
print("However, running Flask directly in this notebook cell is not standard.")
print("The code block above defines the app and saves the template.")
print("You would need to run this in a separate Python script to start the web server.")

# Note: The Flask app is defined but not run within this notebook cell.
# The user would need to copy this code into a .py file and run it separately.
# We cannot run app.run() directly and keep the notebook interactive.
```
